**Selecting X and y to build the ML Model**

```python
#selecting the target variable as profit feature
y = df['Profit(in Thousands)'].to_frame()

#selecting all columns except date and profit as the predictor variables
X = df.drop(['Profit(in Thousands)','Date'], axis=1)
```

#### Task 2: 
#### Build a regression model to predict the profit from the given data and comment your findings on it.

```python
#now we split the dataset into the Dtrain and Dtest in 80-20 ratio since data is small
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
#fitting the Dtrain to sklearn lin reg model
from sklearn.linear_model import LinearRegression
linreg = LinearRegression()
model = linreg.fit(X_train, y_train)
```

```python
#getting the predictions 
y_pred = model.predict(X_test)
```

```python
#cross checking the results 
```

```python
y_test
```

```python
print(y_pred)
```

* we see an exact match of y_test values and y_pred values

#### Checking MSE and R2 Values

```python
#finding out mean-squared error
from sklearn.metrics import mean_squared_error
print("The Mean Squared Error is:",round(mean_squared_error(y_test,y_pred),4))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test,y_pred)
print('r2 score for perfect model is', r2)
```

**Findings**
* Since the dataset is small, we use 80-20 data split for Dtrain-Dtest splitting.
* On fitting the model and prediction, we see that the predictions are very accurate with an MSE of 0.0 and Coefficient of Determination (R2) of 1.0
* MSE of 0 and R2 of 1.0 are characteristic of perfect Linear Regression models. 
* We can infer that since our dataset is small (only of 205 rows) and as there is a high correlation (due to high similarity of distribution of data in each of the columns), the model got trained quite well. 
* However, this might not be the case for large datasets with less correlation for the same model.
